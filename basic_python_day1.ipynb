"""Control flow - if, if else, if elif else"""

```python
age=int(input("enter your age:"))
if age>=18:
  print("you are eligible for voting")
  print("Please vote")
print("Thank you")
```

```python
age=int(input("enter your age:"))
if age>=18:
  print("you are eligible for voting")
  print("Please vote")
else:
  print("you are not eligible for voting")
  print("Please try after",18-age,"years")
print("Thank you")
```

```python
age=int(input("enter your age:"))
if age>=18:
  print("you are eligible for voting")
  print("Please vote")
elif age>=15:
  print("you are eligible for TEMPORARY voting")
  print("Please vote")
else:
  print("you are not eligible for voting")
  print("Please try after",18-age,"years")
print("Thank you")
```

Regular expression

```python
import re
pattern = re.compile("^(\d{2})-(\d{2})-(\d{4})$")
valid = pattern.match("01/01/2000")
if valid:
  print ("Valid date!")
else:
  print("Invalid Date!")
```

```python
str = 'My email is	abc123@google.com, His email is	cde@gmail.com Her mail is sss@dasd.in. New mail is sash.ibm@gmail.com'
emails = re.findall(r'[\w\.-]+@[\w\.-]+', str)
for email in emails:
  print(email)
```
