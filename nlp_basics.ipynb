```python
import spacy
from spacy.lang.en import English
nlp = English()
```

"""Split into tokens"""

```python
doc = nlp("Apple is looking at buying U.K. startup for $1 billion")
for t in doc:
    print(t.text)
```

"""Extract Specific word from the given sentence"""

```python
span= doc[5:7]
print(span.text)
```

"""Print the stop words of Spacy"""

```python
from spacy.lang.en.stop_words import STOP_WORDS
print(STOP_WORDS)
```

Remove Stop Words

```python
filtered_sentence =[]
doc = nlp("Apple is looking at buying U.K. startup for $1 billion")
for word in doc:
    if  word.is_stop == False:
        filtered_sentence.append(word)
print(doc.text)
print(filtered_sentence)
```

"""Part of Speech -- Grammer"""

```python
nlp = spacy.load("en_core_web_sm")
doc = nlp("He drinks a drink")
for token in doc:
    print(token.text,"\t", token.pos_,"\t", token.tag_)
```

"""Extracting the NOUN"""

```python
doc = nlp("The man reading the news is very tall. The tall man is very kind to everyone")
# doc.noun_chunks is a generator that yields spans
for token in doc.noun_chunks:
    print(token.text)
```

"""Root of noun from a tree"""

```python
for token in doc.noun_chunks:
    print(token.root.text)
```

"""Root Head of Noun"""

```python
for token in doc.noun_chunks:
    print(token.root.head.text)
```

"""Dependency Relationship"""

```python
doc = nlp(u"The dog chased the cat")
for word in doc:
    print(word.text, word.dep_)
```

```python
nlp = spacy.load("en_core_web_sm")
text = """In ancient Rome, some neighbors live in three adjacent houses. In the center is the house of Senex, who lives there with wife Domina, son Hero, and several slaves, including head slave Hysterium and the musical's main character Pseudolus. A slave belonging to Hero, Pseudolus wishes to buy, win, or steal his freedom. One of the neighboring houses is owned by Marcus Lycus, who is a buyer and seller of beautiful women; the other belongs to the ancient Erronius, who is abroad searching for his long-lost children (stolen in infancy by pirates). One day, Senex and Domina go on a trip and leave Pseudolus in charge of Hero. Hero confides in Pseudolus that he is in love with the lovely Philia, one of the courtesans in the House of Lycus (albeit still a virgin)."""
doc = nlp(text)
sentence_spans = list(doc.sents)

from spacy import displacy

displacy.render(sentence_spans, style="dep",jupyter = True)
```

```python
from spacy.tokens import Span

nlp = spacy.load("en_core_web_sm")
doc = nlp("fb is hiring a new vice president of global policy")
ents = [(e.text, e.start_char, e.end_char, e.label_) for e in doc.ents]
print('Before', ents)

label = nlp.vocab.strings["ORG"]
fb_ent = Span(doc,0,1, label=label) # create a Span for the new entity
doc.ents = list(doc.ents) + [fb_ent]

ents = [(e.text, e.start_char, e.end_char, e.label_) for e in doc.ents]
print('After', ents)
```

"""Regular Expression Matcher"""

```python
from spacy.matcher import Matcher
from spacy.tokens import Span
matcher = Matcher(nlp.vocab)

print(matcher)

from spacy.matcher import Matcher
from spacy.tokens import Span
matcher = Matcher(nlp.vocab)
```

```python
doc = nlp("I drink a drink")
pattern = [{"TEXT": "drink","POS":"NOUN"}]
matcher.add("Pattern", [pattern], on_match=None)
matches = matcher(doc)
print("Total matches found:", len(matches))

for match_id, start, end in matches:
    print("Match found:", doc[start:end].text)

for match_id, start, end in matches:
    span = doc[start:end]  # The matched span
    print(match_id, start, end, span.text)
```

```python
matcher1 = Matcher(nlp.vocab)
doc = nlp(
    "After making the iOS update you won't notice a radical system-wide "
    "redesign: nothing like the aesthetic upheaval we got with IOS 7. Most of "
    "iOS 11's furniture remains the same as in iOS 10. But you will discover "
    "some tweaks once you delve a little deeper."
)

# pattern for full iOS versions ("iOS 7", "iOS 11", "iOS 10")
pattern = [{"TEXT": "iOS"}, {"IS_DIGIT": True}]

matcher1.add("IOS_VERSION_PATTERN", [pattern], on_match=None)
matches = matcher1(doc)
print("Total matches found:", len(matches))

for match_id, start, end in matches:
    print("Match found:", doc[start:end].text)

for match_id, start, end in matches:
    span = doc[start:end]  # The matched span
    print(match_id, start, end, span.text)
```
